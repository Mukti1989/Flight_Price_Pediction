```python
# importing all required library
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(2)
```

```python
# ESAYMY TRIP

Airline =[]
Dep_time =[]
Arr_time =[]
Durtion =[]
Route =[]
Source =[]
Destination =[]
Stops =[]
Price =[]
Date =[]
```

```python
driver.get('https://flight.easemytrip.com/FlightList/Index?srch=BLR-Bangalore-India|DEL-NewDelhi-India|01/06/2022&px=1-0-0&cbn=0&ar=undefined&isow=true&isdm=true&lng=&CouponCode=&ompAff=&bc=&ISWL=')



time.sleep(3)
```

```python
for i in range(45):
    heading =driver.find_element_by_xpath('//*[@id="jssor_1"]/div/div/div/div/div[2]/div[7]/div[1]/div[1]/div[1]')
    heading.click()
    time.sleep(6)
    
    

    #Airline
    for i in driver.find_elements_by_xpath("//*[@id='ResultDiv']/div/div/div/div/div/div/div/div/div/span[1]"):
        Airline.append(i.text)


    #departure time
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[2]/span[1]'):
        Dep_time.append(i.text)



    #arrival time
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[4]/span[1]'):
        Arr_time.append(i.text)


    #duration in hours
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[3]/span[1]'):
        Durtion.append(i.text)


    #route 
    count=0

    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[3]/span[2]'):
        Stops.append(i.text)
        detail =driver.find_element_by_xpath(f"//*[@id='{count}']")
        detail.click()
        route1 =driver.find_element_by_xpath(f'//*[@id="fd{count}"]/div/div[2]/div[1]/span[1]/span[1]')
        route2 =driver.find_element_by_xpath(f'//*[@id="fd{count}"]/div/div[2]/div[1]/span[1]/span[2]')
        """try:
            route3 =driver.find_element_by_xpath(f'//*[@id="fd{count}"]/div/div[2]/div[2]/span[1]/span[2]')
        except:
            break"""



        str1 = '--->'

        via =route1.text + str1
        route4= via + route2.text
        # route5 =route4 + str1 
        # route6 =route5 + route3.text
        Route.append(route4)


        count +=1

    #source
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div[1]/div[2]/span[2]'):
        Source.append(i.text)
    #destination
    for i in driver.find_elements_by_xpath('//*[@id="ResultDiv"]/div/div/div/div/div/div/div[4]/span[2]'):
        Destination.append(i.text)
    #price    
    for i in driver.find_elements_by_xpath('//*[@class="col-md-8 col-sm-8 col-xs-9 txt-r6-n ng-binding"]'):
        Price.append(i.text)
    #date
    i  = driver.find_elements_by_xpath('//*[@class="col-md-8 col-sm-8 col-xs-9 txt-r6-n ng-binding"]')
    for j in i:
        date1 =driver.find_element_by_xpath('//*[@id="jssor_1"]/div/div/div/div/div[2]/div[6]/div[1]/div[1]/div[1]') 
        Date.append(date1.text)
```
